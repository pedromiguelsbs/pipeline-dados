# Junção de duas bases de dados

##### **1) Lendo os dados JSON**

```python
path_json = '../raw_data/dados_empresaA.json'

import json

with open(path_json, 'r') as file:
    leitura_json = json.load(file)

leitura_json
```

```python
print(type(leitura_json))
print(leitura_json[0])
print(type(leitura_json[0]))
```

##### **2) Lendo os dados CSV**

```python
path_csv = '../raw_data/dados_empresaB.csv'

import csv

leitura_csv = []

#Lista de listas
#with open(dados_csv, 'r') as file:
#    spamreader = csv.reader(file, delimiter=',')
#
#    for linha in spamreader:
#        leitura_csv.append(linha)

with open(path_csv, 'r') as file:
    spamreader = csv.DictReader(file, delimiter=',')

    for linha in spamreader:
        leitura_csv.append(linha)
```

```python
print(type(leitura_csv))
print(leitura_csv[0])
print(type(leitura_csv[0]))
```

##### **3) Analisando colunas**

```python
nomes_colunas_json = list(leitura_json[0].keys())
print("Qtd de colunas:", len(nomes_colunas_json))
nomes_colunas_json
```

```python
nomes_colunas_csv = list(leitura_csv[0].keys())
print("Qtd de colunas:", len(nomes_colunas_csv))
nomes_colunas_csv
```

##### **4) Padronizando colunas: Alterando colunas do CSV**

```python
novo_rotulo = {'Nome do Item' :'Nome do Produto', 
                 'Classificação do Produto': 'Categoria do Produto', 
                 'Valor em Reais (R$)': 'Preço do Produto (R$)',
                 'Quantidade em Estoque': 'Quantidade em Estoque',
                 'Nome da Loja': 'Filial',
                 'Data da Venda': 'Data da Venda'}
novo_rotulo
```

```python
nova_leitura_csv = []

for dicionario in leitura_csv:
    dicionario_novo = {}
    for chave_antiga, valor in dicionario.items():
        dicionario_novo[novo_rotulo[chave_antiga]] = valor
    nova_leitura_csv.append(dicionario_novo)

nova_leitura_csv
```

##### **5) Fusão dos dados**

```python
print("Dados no JSON:", len(leitura_json))
print("Dados no CSV:", len(nova_leitura_csv))
print("Dados totais:", len(leitura_json) + len(nova_leitura_csv))
```

```python
dados_combinados = []
dados_combinados.extend(leitura_json)
dados_combinados.extend(nova_leitura_csv)
```

```python
len(dados_combinados)
```

##### **6) Salvando os dados ***(lista de dicionários)*****

```python
path = '../processed_data/dados_combinados.csv'
colunas = list(dados_combinados[0].keys())

with open(path, 'w') as file:
    writer = csv.DictWriter(file, fieldnames=colunas)
    writer.writeheader()
    writer.writerows(dados_combinados)
```

##### **7) Resolvendo o problema: Coluna faltante + alterando para lista de listas**

```python
colunas 
```

```python
colunas_novo = list(dados_combinados[-1].keys())
colunas_novo
```

```python
dados_combinados_novo = [colunas_novo]

for item in dados_combinados:
    linha = []
    for coluna in colunas_novo:
        linha.append(item.get(coluna, 'Indisponível'))
    dados_combinados_novo.append(linha)
```

##### **8) Salvando os dados ***(lista de listas)*****

```python
path = '../processed_data/dados_combinados.csv'

with open(path, 'w') as file:
    writer = csv.writer(file)
    writer.writerows(dados_combinados_novo)
```
